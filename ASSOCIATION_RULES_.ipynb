```python
unique_combinations = df.groupby('shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil').agg(list).reset_index()
unique_combinations
```

```python
transaction=unique_combinations['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].tolist()
transaction[0]
```

```python
#apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Converting transaction to columns using transaction_encoder
te=TransactionEncoder()
te_ary=te.fit_transform(transaction)
dataset=pd.DataFrame(te_ary,columns=te.columns_)
dataset
```

```python
#Converting dataset into 1-0 encoding
def encode_units(x):
    if x==False:
        return 0
    if x== True:
        return 1
dataset= dataset.applymap(encode_units)
dataset.head(10)
```

```python
frequent_itemsets= apriori(dataset, min_support=0.10, max_len=2, use_colnames=True)
rules= association_rules(frequent_itemsets, metric="confidence",min_threshold=0.20)
rules.head(20)
```

```python
frequent_itemsets= apriori(dataset, min_support=0.01, max_len=2, use_colnames=True)
rules= association_rules(frequent_itemsets, metric="confidence",min_threshold=0)
rules.head(20)
```
