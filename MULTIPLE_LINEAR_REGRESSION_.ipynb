```python
# Import the pandas library for data manipulation and analysis
import pandas as pd

# Load the dataset into a pandas DataFrame
df=pd.read_csv('/content/ToyotaCorolla - MLR.csv')
df
```

```python
# Display a statistical summary of the DataFrame
df.describe()
```

```python
# Drop columns with any missing values from the DataFrame
df.dropna(axis=1,inplace=True)
df
```

```python
# Drop the 'Fuel_Type' column from the DataFrame
df.drop(['Fuel_Type'],axis=1,inplace=True)
df
```

```python
# Import matplotlib for plotting
import matplotlib.pyplot as plt
# Import seaborn for enhanced data visualization
import seaborn as sns
# Loop through each column in the DataFrame to identify and visualize outliers
for i in df.columns:
  Q1=df[i].quantile(0.25) # Calculate the first quartile (25th percentile) for the current column
  Q3=df[i].quantile(0.75) # Calculate the third quartile (75th percentile) for the current column
  IQR=Q3-Q1               # Calculate the interquartile range (IQR) for the current column
  # Calculate the lower limit for identifying outliers
  lower_limit=Q1-1.5*IQR
  # Calculate the upper limit for identifying outliers
  upper_limit=Q3+1.5*IQR
  # Identify outliers in the current column
  outliers=df[(df[i]<lower_limit) | (df[i]>upper_limit)]
  # Plot a boxplot for the current column to visualize outliers
  plt.figure(figsize=(10,5)) # Set the figure size for the plot
  sns.boxplot(df[i])         # Create a boxplot using seaborn for the current column
  plt.show()                 # Display the plot
```

```python
# Generate pair plots to visualize relationships between pairs of numerical variables
sns.pairplot(df)
plt.show()
```

```python
# Plot histograms for each column in the DataFrame
for i in df.columns:
  plt.hist(df[i],bins=10,edgecolor='black') # Create a histogram with 10 bins and black edges
  plt.title(i)                              # Set the title of the plot as the name of the current column
  plt.show()                                # Display the plot
```

```python
# Generate a heatmap to visualize the correlation matrix
sns.heatmap(df.corr(),annot=True,fmt='.2f')# Create a heatmap using seaborn
plt.title('correlation heatmap')# Set the title of the heatmap
plt.show()# Display the plot
```

```python
# Import the train_test_split function from scikit-learn
from sklearn.model_selection import train_test_split
# Define the feature matrix (X) and the target variable (y)
x=df.drop(columns=['Price'])
y=df['Price']
```

```python
# Split the data into training and testing sets (80% training, 20% testing)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(y_test)
print(x_test)
```

**Building a multiple linear regression model**

```python
 # Import the LinearRegression class from scikit-learn
from sklearn.linear_model import LinearRegression
 # Import evaluation metrics
from sklearn.metrics import mean_squared_error,r2_score
# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred=model.predict(x_test)

#Evaluate the model
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

#Printing the Performance
print("Model Performance")
print("Mean squared error:",mse)
print("R2 Score:",r2)
```

```python
# Import Lasso and Ridge regression models from scikit-learn
from sklearn.linear_model import Lasso, Ridge
 # Import cross_val_score for cross-validation
from sklearn.model_selection import cross_val_score

#Lasso Regression
# Create a Lasso regression model with alpha=0.1
lasso=Lasso(alpha=0.1)
lasso.fit(x_train,y_train) # Fit the Lasso model on the training data
y_pred_lasso=lasso.predict(x_test)# Make predictions on the test data using the Lasso model
# Calculate the mean squared error for the Lasso model
mse_lasso=mean_squared_error(y_test,y_pred_lasso)
# Calculate the R-squared value for the Lasso model
r2_lasso=r2_score(y_test,y_pred_lasso)
#Printing the Performance
print("Lasso Regression performance")
print("Mean squared error:",mse_lasso)
print("R-squared",r2_lasso)

#Ridge Regression
# Create a Ridge regression model with alpha=0.1
ridge=Ridge(alpha=0.1)
ridge.fit(x_train,y_train)# Fit the Ridge model on the training data
y_pred_ridge=ridge.predict(x_test)# Make predictions on the test data using the Ridge model
# Calculate the mean squared error for the Ridge model
mse_ridge=mean_squared_error(y_test,y_pred_ridge)
# Calculate the R-squared value for the Ridge model
r2_ridge=r2_score(y_test,y_pred_ridge)
#Printing the performance
print("\nRidge Regression performance")
print("Mean squared error:",mse_lasso)
print("R-squared",r2_lasso)
```
