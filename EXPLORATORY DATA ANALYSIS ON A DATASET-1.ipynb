**EXPLORATORY DATA ANALYSIS **

```python
import pandas as pd  # Import the pandas library for data manipulation and analysis

# Load the dataset into a pandas DataFrame
df = pd.read_csv('/content/Cardiotocographic.csv')

# Display the entire DataFrame
df
```

```python
# Display a statistical summary of the DataFrame
df.describe()
```

```python
# Drop rows with any missing values from the DataFrame
df.dropna(inplace=True)

# Display the DataFrame after dropping missing values
df
```

```python
import matplotlib.pyplot as plt  # Importing matplotlib for plotting
import seaborn as sns  # Importing seaborn for enhanced data visualization

# Loop through each column in the DataFrame to identify and visualize outliers
for i in df.columns:
    Q1 = df[i].quantile(0.25)  # Calculate the first quartile (25th percentile) for the current column
    Q3 = df[i].quantile(0.75)  # Calculate the third quartile (75th percentile) for the current column
    IQR = Q3 - Q1  # Calculate the interquartile range (IQR) for the current column

    # Calculate the lower limit for identifying outliers
    lower_limit = Q1 - 1.5 * IQR
    # Calculate the upper limit for identifying outliers
    upper_limit = Q3 + 1.5 * IQR

    # Identify outliers in the current column
    outliers = df[(df[i] < lower_limit) | (df[i] > upper_limit)]
    print(f'outlier of {i}={outliers}')  # Print the outliers for the current column

    # Plot a boxplot for the current column to visualize outliers
    plt.figure(figsize=(10, 5))  # Set the figure size for the plot
    sns.boxplot(df[i])  # Create a boxplot using seaborn for the current column
    plt.show()  # Display the plot
```

**SUMMARY**
Highlighting interesting findings
Interesting Findings:
1. If any column has a mean significantly higher or lower than the median, it may indicate skewness in the data.
2. High standard deviation relative to the mean suggests high variability.
3. Large interquartile ranges indicate substantial spread in the middle 50% of the data.
4. Presence of outliers as seen in the box plots indicates points that are far from the rest of the data.

```python
# Iterate through each column in the DataFrame
for i in df.columns:
    # Plot a histogram for the current column
    plt.hist(df[i], bins=10, edgecolor='black')  # Create a histogram with 10 bins and black edges
    plt.title(i)  # Set the title of the plot as the name of the current column
    plt.xlabel(i)  # Set the x-axis label as the name of the current column
    plt.ylabel('count')  # Set the y-axis label as 'count'
    plt.show()  # Display the plot
```

```python
# Generate scatter plots for pairs of numerical variables
sns.pairplot(df)
plt.show()
```

```python
# Calculate the correlation matrix
# Generate a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

The pandas library is imported as 'pd' to manipulate and analyze data.

**Data Cleaning and Preparation:**

The entire DataFrame is displayed to inspect the dataset.

Rows with missing values are dropped using df.dropna(inplace=True) to ensure a clean dataset.

**Outlier Detection and Visualization:**

For each column in the DataFrame, outliers are detected using the interquartile range (IQR) method.
