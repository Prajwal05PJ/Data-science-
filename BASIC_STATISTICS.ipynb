```python
import seaborn as sns
from scipy.stats import skew
skewness = skew(df['Volume'].dropna())
print("Skewness of Avg Price:",skewness)
Q1 = df['Volume'].quantile(0.25)
Q3 = df['Volume'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Volume'] < lower_bound) | (df['Volume'] > upper_bound)]
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower bound for outliers: {lower_bound}")
print(f"Upper bound for outliers: {upper_bound}")
print(f"Number of outliers in {'Volume'}: {outliers.shape[0]}")
print("Outliers:")
print(outliers['Volume'])
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Volume'])
plt.title('Volume')
plt.xlabel('Volume')
plt.show()
```

```python
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']


for i in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[i], order=df[i].value_counts().index)
    plt.title(f'Frequency of {i}')
    plt.xlabel('Count')
    plt.ylabel(i)
    plt.show()
```

```python
df1 = df.copy() #using the formula z=x-mu/sigma
df1['Volume'] = (df['Volume'] - df['Volume'].mean()) / df['Volume'].std()
df1['Avg Price'] = (df['Avg Price'] - df['Avg Price'].mean()) / df['Avg Price'].std()
df1['Total Sales Value'] = (df['Total Sales Value'] - df['Total Sales Value'].mean()) / df['Total Sales Value'].std()
df1['Discount Rate (%)'] = (df['Discount Rate (%)'] - df['Discount Rate (%)'].mean()) / df['Discount Rate (%)'].std()
df1['Discount Amount'] = (df['Discount Amount'] - df['Discount Amount'].mean()) / df['Discount Amount'].std()
df1['Net Sales Value'] = (df['Net Sales Value'] - df['Net Sales Value'].mean()) / df['Net Sales Value'].std()
```
